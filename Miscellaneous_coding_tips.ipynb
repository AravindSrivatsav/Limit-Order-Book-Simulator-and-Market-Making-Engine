1) How to add the custom objects to the heap with the criteria of objects attributes

```python
# Define the lt (less than) custom method inside the class. then you can normally heappush.

def __lt__(self, other):
        # Sort priority ASCENDING
        if self.priority != other.priority:
            return self.priority < other.priority
        # Sort label DESCENDING if priorities are equal
        return self.label > other.label
```

2. How to use f string.

```python
name = "Alice"
age = 30
pi = 3.14159

# Using f-strings with a 2-decimal restriction for pi
print(f"Hello {name}, you are {age} years old. Pi is {pi:.2f}")
# Output: Hello Alice, you are 30 years old. Pi is 3.14
```

4. enumerate() function usage

```python
fruits = ["apple", "banana", "cherry"]

# Enumerate returns a pair: (index, item)
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")
```

8. How to indentify a custon object and verify it's class.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

# Create a custom object
my_car = Car("Toyota")

# Check the custom object
if isinstance(my_car, Car):
    print("Yes, this is a Car object!")

# Returns True only if it matches the class exactly
if type(my_car) is Car:
    print("This is exactly a Car object.")
```

9. Use super().__init__(...) in subclasses to initialize the parent part.

```python
class Animal:
    def __init__(self, name):
        self.name = name


class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Call Animal's constructor
        self.breed = breed


d = Dog("Tommy", "Labrador")

print(d.name)
print(d.breed)
```
